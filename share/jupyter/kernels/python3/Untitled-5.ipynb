```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('oil_sales_assignment_dataset.csv')
df
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True, color='#343434')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count') 
plt.show()
```

```python
plt.figure(figsize=(7,7))

df['price_bracket'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Price Bracket Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x='city',
    order=df['city'].value_counts().index
)

plt.xticks(rotation=90)
plt.title("Count of Records by City")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='volume_sales',
    y='value_sales',
    hue='price_bracket'
)

plt.title("Volume Sales vs Value Sales")
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.violinplot(
    data=df,
    x='price_bracket',
    y='average_price',
    palette='Set2'
)

plt.title("Average Price by Price Bracket")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.kdeplot(
    data=df,
    x='average_price',
    fill=True
)

plt.title("KDE Plot of Average Price")
plt.show()
```

```python
g = sns.FacetGrid(
    df,
    col='price_bracket',
    height=4
)

g.map(sns.histplot, 'average_price')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(
    data=df,
    x='price_bracket',
    y='average_price',
    hue='price_bracket',
    palette='Set2'
)

plt.title("Swarm Plot of Average Price by Price Bracket")
plt.xlabel("Price Bracket")
plt.ylabel("Average Price")
plt.legend(title="Price Bracket")
plt.show()
```

```python
df = pd.read_csv("oil_sales_assignment_dataset.csv")

sns.pairplot(
    df,
    vars=['value_sales', 'volume_sales', 'average_price'],
    hue='price_bracket',
    diag_kind='kde',
    palette='Set2'
)

plt.show()
```
