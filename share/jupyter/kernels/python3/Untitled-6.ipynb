```python
# Regression Models
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
df=pd.read_csv("data.csv")

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Missing Values:", df.isnull().sum())
print("duplicate Rows:", df.duplicated().sum())
```

```python
df=df.drop(['date','street','statezip','country'],axis=1)
```

```python
# handle missing values
for col in df.select_dtypes(['int64','float']).columns:
    df[col].fillna(df[col].median(), inplace=True)
for col in df.select_dtypes(['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)    
```

```python
# encode categorical variables
label_encoders = {}
for col in df.select_dtypes(['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# features selection
target_column = 'price'
X=df.drop(target_column, axis=1)
y=df[target_column]
```

```python
#Features scalling
scalar=StandardScaler()
X_scaled=scalar.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print("MAE:",mae)
    print("MSE:",mse)
    print("RMSE:",rmse)
    print("R² Score:",r2)
```

```python
# linear regression
#1.linear regression
print("\n===== Linear Regression =====")
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#2. Muliple linear regression

print("\n===== Multiple Linear Regression =====")
mlr=LinearRegression()
mlr.fit(x_train,y_train)
y_pred=mlr.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures
# Polynomial Regression
print("\n=== Polynomial Regression ===")

poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

lr = LinearRegression()
lr.fit(x_train_poly, y_train)

y_pred = lr.predict(x_test_poly)

evaluate_model(y_test, y_pred)
```

```python
#ridge regression
print("\n===== Ridge Regression =====")
ridge=Ridge(alpha=1.0)
ridge.fit(x_train,y_train)
y_pred=ridge.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
print("\n===== Lasso Regression =====")
lasso=Lasso(alpha=0.1)
lasso.fit(x_train,y_train)
y_pred=lasso.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#Elastic Net Regression
print("\n===== Elastic Net Regression =====")
elastic_net=ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(x_train,y_train)
y_pred=elastic_net.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#Decision tree
print("\n===== Decision Tree =====")
dt=DecisionTreeRegressor(max_depth=5)
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#Random Forest
print("\n===== Random Forest =====")
rf=RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#support vector regrssor
print("\n===== Support Vector Regressor =====")
svr = SVR(kernel='rbf', C=100, gamma=0.1)
y_pred = svr.fit(x_train, y_train).predict(x_test)
evaluate_model(y_test, y_pred)
          
```

```python
# k nearest neighbors regression
print("\n===== K-Nearest Neighbors Regression =====")
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
evaluate_model(y_test, y_pred)
```
