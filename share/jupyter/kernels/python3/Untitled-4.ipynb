```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
df
```

```python
cars=df['owner'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(cars, labels=['First Owner','Second owner','Third Owner','Fourth & Above Owner','Test Drive Car'],
        autopct='%1.1f%%',shadow=True, startangle=140,colors=['lightcoral','red','blue','green','orange'])
plt.title('Survival Rate on TITANIC')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True, color='#343434')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count') 
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type',data=df,palette='viridis')
plt.title('selling_type count plot')

plt.xlabel('seller type')
plt.ylabel('count')
plt.show
```
