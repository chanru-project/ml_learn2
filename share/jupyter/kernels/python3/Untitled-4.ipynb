```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
df
```

```python
cars=df['owner'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(cars, labels=['First Owner','Second owner','Third Owner','Fourth & Above Owner','Test Drive Car'],
        autopct='%1.1f%%',shadow=True, startangle=140,colors=['lightcoral','red','blue','green','orange'])
plt.title('Survival Rate on TITANIC')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True, color='#343434')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count') 
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type',data=df,palette='viridis')
plt.title('selling_type count plot')

plt.xlabel('seller type')
plt.ylabel('count')
plt.show
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='year',y='km_driven',data=df,hue='owner',style='seller_type',palette='rainbow')
plt.title('Scatter plot of year vs km_driven with owner and seller_type ')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.legend(title='km_driven')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='year',y='km_driven',data=df,hue='owner',style='seller_type',palette='rainbow')
plt.title('Scatter plot of year vs km_driven with owner and seller_type ')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.legend(title='km_driven')
plt.show()
```

```python
df.info()
```

```python
num=df[['year','km_driven','selling_price']].corr()
num
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(num, annot=True, cmap='rainbow',center=0)
plt.title('Correlation Hearmap of Car Dataset')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(
    x='fuel',
    y='year',
    hue='seller_type',
    data=df,
    palette='Set2'
)
plt.title('seller_type vs fuel vs year')
plt.xlabel('fuel')
plt.ylabel('year')
plt.legend(title='suruvived')
plt.show()
```

```python
sns.kdeplot(data=df['year'].dropna(), color='red',shade=True)
plt.title('KDE plot for year')
plt.xlabel('year')
plt.ylabel('Density')
plt.show()
```

```python
g=sns.FacetGrid(df, col='seller_type', row='owner', margin_titles=True)
g.map(sns.histplot,'year',bins=20,color='darkblue')
g.fig.suptitle('Histplot of year with seller_type and owner',y=1.05)
plt.show()
```

```python
plt.figure(figsize=(5, 5))

sns.swarmplot(
    x='fuel',
    y='km_driven',
    hue='transmission',
    data=df,
    palette='Set1',
    dodge=True
)

plt.title('Transmission Distribution by Fuel and km_driven (Swarm Plot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```

```python
sns.pairplot(
    df[['year', 'selling_price', 'km_driven', 'transmission']],
    hue='transmission',
    palette='husl',
    plot_kws={'alpha': 0.6},
    diag_kind='kde'
)

plt.suptitle('Pair plot of titanic variables', y=1.02)
plt.show()
```
